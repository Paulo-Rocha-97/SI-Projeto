```python
def policy(state):
    
    txt=input('\nAction to be made ( 1 - Hit ; O - Stand) :')
    action= int(txt)
        
    return action
```

```python
import gym
import scipy.io as sio

env=gym.make('Blackjack-v0')

myDictionary = {}
in_state =  []
Action_done = []
Final_state = []
Reward_vec=[]
Data_inter=[]
Data_action=[]


txt = input('Numero de jogos pretendidos:')
Numero = int(txt)

for i in range(Numero):
    
    print('\nNew Game \n')
    
    state = env.reset()
    
    L=list(state)    
    in_state.append(L)

    print('Initial score: %d' % (state[0]))
    print('Dealer Face up card %d' % (state[1]))

    done = False
    
    cont = 0
    
    Action_done = []
    Inter_state = []
    
    while not done: 

        action = policy(state)

        state, reward, done, info = env.step(action)

        print('\nNew_score:',state[0])
        
        Action_done.append(action)
        
        L=list(state)
        Inter_state.append( L )
        
        cont = cont + 1

    print('\nFinal result:\n  Your score: %d\n  Dealer\'s score: %d\n'% (state[0],state[2]))
    
    if reward == 1:
    
        print('You win!!')
    
    elif reward == 0:
    
        print('You tie!!')
        
    elif reward == -1:

        print('You lose!!')
        
    Final_state.append(state) 
    Reward_vec.append(reward)
    Data_inter.append(Inter_state)
    Data_action.append(Action_done)

filename = input('Filename:')
    
name = filename+'.mat' 

myDictionary['In_sate'] = in_state
myDictionary['Inter_state'] = Data_inter
myDictionary['Action'] = Data_action
myDictionary['Fi_state'] = Final_state
myDictionary['Reward'] = Reward_vec


    
sio.savemat(name, myDictionary)  


env.close()
```
